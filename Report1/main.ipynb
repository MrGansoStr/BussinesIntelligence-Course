```python
def obtener_segunda_parte(cadena):
    try:
        partes = cadena.split('-')
        # print(len(partes))
        if len(partes) == 2:
            # ga = ""
            # ga.replace(" ")
            return partes[1].replace(" ", "")
        else:
            return partes[0].replace(" ", "")
    except Exception as e:
        print(str(e), "aaaaaaaaaaa")
    # return partes[1].strip() if len(partes) > 1 else cadena.strip()
```

```python
# cadena = "0000002990 - JULI"
# segunda_parte = cadena.split('-')[1].strip()
final_df.loc[:,"EESS_ATENCION_Fixed"] = final_df["EESS ATENCION"].apply(obtener_segunda_parte)
```

```python
pd.unique(final_df["EESS_ATENCION_Fixed"])
```

```python
final_df.loc[:, "EESS_ATENCION_Fixed_Num"] = final_df["EESS_ATENCION_Fixed"].astype('category').cat.codes
```

```python
final_df
```

```python
final_df.loc[:, "TipoAtencion_Num"] = final_df["TipoAtencion"].map({'AMBULATORIO': 0, 'EMERGENCIA': 1, 'REFERIDO': 2})
```

```python
final_df
```

```python
final_df.loc[:, "Sexo_num"] = final_df["Sexo"].apply(lambda x: 1 if x in "M" else 0)
# 1 M
# 0 F
```

```python
import matplotlib.pyplot as plt
```

```python
freq_table = final_df.groupby(['EESS_ATENCION_Fixed', 'dg_fixed']).size().unstack(fill_value=0)

# Crear el gráfico de barras
freq_table.plot(kind='bar', stacked=True)
plt.xlabel('EESS atencion')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de dg_aux por EESS atencion')
plt.legend(title='dg_aux')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data_dg_aux_0 = final_df[final_df['dg_fixed'] == 0]
data_dg_aux_1 = final_df[final_df['dg_fixed'] == 1]

# Calcular la frecuencia de EESS atencion para cada valor de dg_aux
freq_table_0 = data_dg_aux_0['EESS_ATENCION_Fixed'].value_counts()
freq_table_1 = data_dg_aux_1['EESS_ATENCION_Fixed'].value_counts()

# Crear el primer gráfico para dg_aux = 0
plt.figure(figsize=(10, 5))
bars = plt.bar(freq_table_0.index, freq_table_0.values, color='blue')

# Agregar etiquetas de frecuencia en cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

plt.xlabel('EESS atencion')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de EESS atencion para dg_aux = 0')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# Crear el segundo gráfico para dg_aux = 1
```

```python
plt.figure(figsize=(10, 5))
bars = plt.bar(freq_table_1.index, freq_table_1.values, color='green')

# Agregar etiquetas de frecuencia en cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

plt.xlabel('EESS atencion')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de EESS atencion para dg_aux = 1')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
